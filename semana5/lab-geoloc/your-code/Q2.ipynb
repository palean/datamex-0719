En esta parte, la tarea consiste en llamar a MongoDB para realizar geoqueries y representar los datos geoespaciales con folium, mostrando en un mapa la geolocalizacion de la nueva oficina.

```python
import pymongo
from pymongo import MongoClient
import folium
from folium import plugins
import pandas as pd
from bson.json_util import loads
```

```python
# Making the connection with mongoDB

client = MongoClient('mongodb://localhost:27017/')
db = client.db_comp
```

```python
# Due my mongoDB compass didn't upload the 'geop.json', here is the loading of the data

db['geop'].insert_many([loads(one) for one in open('geop.json','r')][0]);
```

```python
# Converting data into a df

geop_df = pd.read_json('geop.json')
```

```python
# Creating the location field

db.geop.create_index([('Geolocation', pymongo.GEOSPHERE)])
```

```python
# Applying the findNear function to the entries of the Geolocation column

def findNear(location, ratio=2000):
    return db.geop.find(
                        {'Geolocation':
                             {'$near':{'$geometry':location, '$maxDistance':ratio}}
                        }
                        )
# Creating a new column with the nearest point office
geop_df['nearest'] = geop_df['Geolocation'].apply(lambda x: list(findNear(x)))
```

```python
geop_df.shape
```

```python
points = folium.Map([37.446823, -122.161523], zoom_start=11)  #initializing a point

#This map is showing 1416 offices from possible clients where it can be found the best neighbour ratio by enterprise category.
for p in geop_df.index:
    folium.Marker([geop_df.loc[p,'latitude'], geop_df.loc[p,'longitude']],
                popup=geop_df.loc[p,'name']).add_to(points)

points
```
