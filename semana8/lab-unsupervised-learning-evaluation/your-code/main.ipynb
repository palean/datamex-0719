```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.rcParams['figure.figsize']=(12,6)
```

```python
from sklearn import datasets

data = datasets.load_wine()

X = pd.DataFrame(data["data"], columns=data["feature_names"])
y = pd.Series(data["target"])

X.head()
```

## 1. Train a KMeans clustering model on the data set using 8 clusters and compute the silhouette score for the model.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

from sklearn.preprocessing import StandardScaler
```

```python
#kmeans = KMeans()                                  # Creating the KMeans model classification

X = StandardScaler().fit_transform(X)               # Normalization of X
kmeans = KMeans(n_clusters=8)                       # Defining model with clusters numbers
kmeans.fit(X)                                       # Fitting the model to the scaled X
y_hat = kmeans.predict(X)                           # Predicting based on X

print(f'Silhouette score for 8 clusters:{round(silhouette_score(X,y_hat),3)}')
```

## 2. Train a KMeans clustering model on the data set using 5 clusters and compute the silhouette score for the model.

```python
kmeans = KMeans(n_clusters=5)                       # Defining model and clusters
kmeans.fit(X)                                       # Fitting the model to X
y_hat = kmeans.predict(X)                           # Predicting based on X

print(f'Silhouette score for 5 clusters:{round(silhouette_score(X,y_hat),3)}')
```

## 3. Train a KMeans clustering model on the data set using 3 clusters and compute the silhouette score for the model.

```python
kmeans = KMeans(n_clusters=3)                       # Defining model and clusters
kmeans.fit(X)                                       # Fitting the model to X
y_hat = kmeans.predict(X)                           # Predicting based on X

print(f'Silhouette score for 3 clusters:{round(silhouette_score(X,y_hat),3)}')
```

```python
#It seems that the silhouette score does not catches the entire density of the clusters and how well separated are their points.
# Just with 3 clusters it reached a silhouette of 28.5%.
```

## 4. Use elbow curve visualizations to see if you can determine the best number of clusters to use.

The Yellowbrick library has 3 metrics that you can plot using the `metric` parameter:

- **distortion**: mean sum of squared distances to centers
- **silhouette**: mean ratio of intra-cluster and nearest-cluster distance
- **calinski_harabaz**: ratio of within to between cluster dispersion

```python
from yellowbrick.cluster import KElbowVisualizer

model = KMeans()

visualizer = KElbowVisualizer(model,k=(2,12),metric='distortion')
visualizer.fit(X)
visualizer.poof()

# In this model, the distorsion score is directly proporcional to the number of clusters
```

```python
model = KMeans() 
visualizer = KElbowVisualizer(model,k=(2,12),metric='silhouette')
visualizer.fit(X)
visualizer.poof()

#while the silhouette decrease with the number of clusters.
```

```python
model = KMeans()
visualizer = KElbowVisualizer(model,k=(2,12),metric='calinski_harabaz')
visualizer.fit(X)
visualizer.poof();

# The calinski_harabasz score shows how from the third cluster the ratio of dispersion between and within clusters decrease steeply.
```

## 6. Create and plot a scatter matrix showing how the clusters are grouped across all the different combinations of variables in the data.

Use the model and number of clusters that returned the best result above.

```python
df = pd.DataFrame(X,columns=data['feature_names'])

model = KMeans(n_clusters=3)
model.fit(X)
df['Cluster']=model.predict(X)

sns.pairplot(df,hue='Cluster',vars=df.columns[:-1])
```

## 7. Apply a PCA transform and plot the first two principle components with the plot point colors determined by cluster.

```python
from sklearn.decomposition import PCA
```

```python
# Showing how is the % of variance by each number of components

pca=PCA()
pca.fit(X)
pca.explained_variance_ratio_

plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('% of Variance')
plt.ylim([0,1])
plt.show();
```

```python
# From the graphic above,the model can be explained just until the eighth component by 90%.

np.cumsum(pca.explained_variance_ratio_) * 100
```

```python
# So, the second PCA component explains the 55.4% of the model.

pca_x = pca.transform(X)

pca_df = pd.DataFrame([pca_x[:,0],pca_x[:,1]]).T
pca_df.columns=['PCA1','PCA2']
pca_df = pd.concat([pca_df, df['Cluster']], axis=1)

sns.lmplot('PCA1','PCA2', data=pca_df, hue='Cluster',fit_reg=False,palette='viridis')
```

## 8. Generate a series of t-SNE plots showing the clusters at a variety of perplexities.

```python
from sklearn.manifold import TSNE

perpl = [10,30,50,100,200]

fig, ax = plt.subplots(1,5,figsize=(15,7))

for i,p in enumerate(perpl):
    tsne = TSNE(perplexity=p)
    x_tsne = tsne.fit_transform(X)
    ax[i].scatter(x_tsne[:, 0], x_tsne[:, 1], c=y, cmap='viridis')
    ax[i].set_title("Perplexity = {}".format(p))

plt.show()    
```
